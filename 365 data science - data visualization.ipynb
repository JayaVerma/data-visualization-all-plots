# Stacked area chart

```python
df_fuel_engine_type = pd.read_csv('C:/Users/HP/Desktop/ML projects/Data_Science_365/Data visualization/Section 5/1_Stacked Area Chart - Introduction - General Theory and Dataset/Course notes/stacked_area_chart_data.csv')
```

```python
df_fuel_engine_type
```

```python
colors = ['#011638', '#7e2987','ef2026']
labels = ['Diesel','Petrol','Gas']
#sns.set_style('white')
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_type['Year'],
             df_fuel_engine_type['Diesel'],
             df_fuel_engine_type['Petrol'],
             df_fuel_engine_type['Gas'],
             colors = colors,
             edgecolor = 'none')
plt.xticks(df_fuel_engine_type['Year'], rotation=45)
plt.legend(labels = labels, loc='upper left', fontsize=15)
plt.ylabel("Numbers of cars", fontsize=15)
plt.title("POPULARITY OF ENGINE FUEL TYPES(1982-2016)", 
          fontsize=20, fontweight='bold')
sns.despine()
plt.show()
```

#### IMP points:-
1. Stacked area charts are great for spotting trends, but only work well with a few categories.
2. Stacked area charts are used to discover trends and patterns in the data
3. follows development of several different variables over time
4. y axis of the chart should began with zero



### What makes a good area chart?

1. when we have time date
2. when we have to compare volume among the categories
3. when we have atleast 3 categories
4. when the variables are ordered and numerical
5. when we have no more than 6 -7 categories

# Histogram

1. Shows the distribution of a numeric variable
2. the variable ranges pf values is split into intervals, represented by different bins
3. the heigh of the bins shows the numbers of observations with in an interval

Histograms are one of the easiestt and the most useful tool to undersatand numerical data.

```python
df_real_estate = pd.read_csv("C:/Users/HP/Desktop/ML projects/Data_Science_365/Data visualization/Section 7/1_Histogram - Introduction - General Theory and Dataset/Course notes/histogram_data.csv")
```

```python
df_real_estate
```

```python
plt.figure(figsize=(8,6))
plt.hist(df_real_estate["Price"],
        bins =8,
        color = '#108A99')
plt.title('Distribution of real Estate Price', fontsize=15, fontweight='bold')
plt.xlabel("Price in (000' $)")
plt.ylabel("Number of properties")
sns.despine()
plt.show()
```

# Scatter plot

1. Can display a large number of points
2. each observation is a point on that scatter
3. Our data is not aggregated

- Can reveal relationships like correlation within the features
- The concentration of observations and outliers in the data is easily visible
- A vital part of the exploratory data process for many statistical models


```python
plt.figure(figsize=(12,8))
scatter = plt.scatter(df_real_estate['Area (ft.)'],
           df_real_estate['Price'],
           alpha=0.6,
           c= df_real_estate['Building Type'],
           cmap = 'viridis')
plt.legend(*scatter.legend_elements(),
          loc = 'upper left',
          title = 'Building Type')
plt.title("Relationship between Area and Price Of Califonia Real Estates",
         fontsize=15,
         weight='bold')
plt.xlabel('Area (sq. ft.)', weight='bold')
plt.ylabel("Price (ooo's of $)", weight='bold')
sns.despine()
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(df_real_estate['Price'],
                df_real_estate['Area (ft.)'],
           hue = df_real_estate['Building Type'],
            palette = ['black', 'darkblue', 'purple', 'pink', 'skyblue'],
           s=100)
plt.title("Relationship between Area and Price Of Califonia Real Estates",
         fontsize=15,
         weight='bold')
plt.xlabel('Area (sq. ft.)', weight='bold')
plt.ylabel("Price (ooo's of $)", weight='bold')
sns.despine()
plt.show()
```

# Regression plot

```python
df_ad_expenditure = pd.read_csv('C:/Users/HP/Desktop/ML projects/Data_Science_365/Data visualization/Section 9/1_Regression Plot - Introduction - General Theory and Dataset/Course notes/scatter_plot_ii.csv')
```

```python
df_ad_expenditure
```

```python
#plt.figure(figsize=(12,8))
sns.set(rc = {'figure.figsize': (9,6)})
sns.regplot(x = 'Budget',
           y = 'Sales',
           data = df_ad_expenditure,
           scatter_kws = {'color':'k'},
           line_kws = {'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)", fontsize=15)
plt.ylabel("Sales in (000's Units)",fontsize=15)
plt.title("Effect of Expenditure on the sales", fontsize=20, weight= 'bold')
plt.show()
```

```python
#plt.figure(figsize=(12,8))
#sns.set(rc = {'figure.figsize': (9,6)})
sns.lmplot(x = 'Budget',
           y = 'Sales',
           data = df_ad_expenditure,
           height =10,
           scatter_kws = {'color':'k'},
           line_kws = {'color':'red'})
plt.xlabel("Ad Expenditure in (000's $)", fontsize=15)
plt.ylabel("Sales in (000's Units)",fontsize=15)
plt.title("Effect of Expenditure on the sales", fontsize=20, weight= 'bold')
plt.show()
```
